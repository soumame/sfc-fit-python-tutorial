### Assignment to variables

* A *variable* is a named box that contains data.
  * Characters that can be used in names are alphabets, numbers, and underscores. However, the first letter must be other than a number.
* Storing a value in a variable is called *assignment*.
* To assign a value to a variable, write `variable name = value`.
* Once assigned, then writing the variable name is the same as writing the value that the variable remembers.

```python
x = 1
print(x)
```

* When you assign a new value, the previously assigned value disappears.

```python
x = 1
x = 2
print(x)
```

* You can assign not only numbers but also other data such as strings to variables.

```python
x = 'Hello, world!'
print(x)
```

* A name that has never been assigned will result in an error because it will not be recognized as a variable.

```python
print(foo)
```

### Variables and expressions

* You can use variable names in an expression.

```python
x = 2
y = x + 1
z = x * y
print(z)
```

* The variable you want to assign can be used in the expression, such as `x = x + 1`.
* The current value of the variable is used to calculate, and the result will be stored as the new value of the variable.

```python
x = 1
x = x + 1
x = x + 1
x = x + 1
print(x)
```

## Functions

### Function call

* A function is something that takes data, does some processing, and returns another piece of data.
* If there is a form of `function name (data)` in the expression, the function is executed. This is called a *function call*.
 * The inside of the parenthese is called *argument* or *parameter*, which is the data received by the function.
 * The data returned by the function is called *return value*.
 * The ` function name (data) ` part in an expression is replaced with the return value, and the next calculation proceeds.
* For example, `len` is a function that takes string data, counts how many characters it has, and returns the number of characters.


```python
print(len('Hello, world!'))
```

```python
print(len('Hello, ') + len('world!'))
```

* If there are two or more arguments, separate them with commas.
* For example, the `max` function is a function that takes several numbers as arguments, finds the largest one, and returns it.

```python
print(max(3,5,1,2,4))
```

### Library functions

* `len` and` max` are built into Python from the beginning. Such functions are called *built-in functions*.
* Functions that are added as needed are called *library functions*.
 * When using a library function, you need to specify the location where the function is stored (called *module*) with `import` and load it.
 * For example, the `sqrt` function that calculates the square root is in a module called `math`.
 * When calling the `sqrt` function, add the module name `math` and a period to form `math.sqrt (data)`.

```python
import math
print(math.sqrt(2))
```

## Data types

* The data handled by Python is divided into several groups. These are called *types*.
* Numbers and strings are different types. Numbers are further divided into integers and floating-point numbers.
 * üîó[Floating Point Numbers - The Floating-Point Guide](https://floating-point-gui.de/formats/fp/)
* If the types are different, different processing will be performed even if they look the same at first glance.
 * For example, 1 is an integer, 1.0 is a floating-point number, and '1' is a string.
 * The `+` operator means 'addition' for integers and floating-point numbers while 'join' for strings.

```python
print( 1 + 2 )
```

```python
print( '1' + '2' )
```

* If you try to calculate by mixing different types of data, you will get a TypeError.
 * However, if you mix integers and floating-point numbers, integers are automatically converted to floating-point numbers and calculated.

```python
print( 1 + '2' )
```

* You can convert data of one type to data of another type.
 * The `int` function takes a string and returns the integer that the string represents.
 * The `float` function takes a string and returns the floating point number that the string represents.
 * The `str` function takes a number and returns the string that represents it.

```python
print( 1 + int('2') )
```

```python
print( str(1) + '2' )
```
