# Lesson10 Conditional

* A program consists of multiple sentences.
* As a general rule, statements are executed in order from the top, but you can specify conditions to execute or not execute a statement.
* This is called *conditional branching*.

## How to write conditions

* An expression whose value is a boolean value is used to decide whether to execute statements.


### Boolean values

* Boolean values â€‹â€‹are a type of data, and there are only two values â€‹â€‹`True` and` False`.

````python
print(True)
print(False)
````

### Comparison operators

* An operator that compares two data and returns a boolean value is called a *comparison operator*.
* For example, an operator that returns `True` when the values â€‹â€‹of two expressions are equal and` False` when they are not equal is `==`.

````python
print(1==1)
print(1==2)
````

* There are other operators such as magnitude comparison.
 * ğŸ”—[Python - Basic Operators - Tutorialspoint](https://www.tutorialspoint.com/python/python_basic_operators.htm)

````python
x = 3
print(x>2)
print(x<2)
print(x>=4)
print(x<=4)
````

## How to write if-statement

* Use an *if-statement* to separate cases.
* Its form differs depending on how many cases you divide.

ğŸ”—[Python if, if...else, if...elif...else and Nested if Statement - Programiz](https://www.programiz.com/python-programming/if-elif-else)

### When the number of cases is one

```
if condition:
  statements
```
* Performs statements when the condition is satisfied (when the value of the expression is `True`).
* If the condition is not satisfied (when the value of the expression is `False`), nothing is done.
* The following figure illustrates it.
  * This kind of chart is called a *flow chart*.

![](https://itclass.sfc.keio.ac.jp/text/info1/image/IfThen-en.png)


* You can write as many statements as you like in the place of 'statements'.
* When the condition is not met, statements that have more whitespaces at the beginning of the line than the `if` line (deeper indentation than `if`) are skipped.

````python
if 1 == 1 :
  print('These statements will be')
  print('executed when the conditon is satisfied.')
print('This is the next statement.')
````

````python
if 1 == 2 :
  print('These statements will be')
  print('executed when the conditon is satisfied.')
print('This is the next statement.')
````

### When the number of cases is two

```
if condition :
  statements 1
else :
  statements 2
```
* Performs statements 1 when the condition is satisfied (when the value of the expression is `True`).
* Performs statements 2 when the condition is not satisfied (when the value of the expression is `False`).
* The following figure illustrates it.

![](https://itclass.sfc.keio.ac.jp/text/info1/image/IfThenElse-en.png)

* The indentation of the `if` line and the `else` line must be the same.
* The indentation of statements 1 and statements 2 must be deeper than the `if` and `else` lines.

````python
if 1 == 1 :
  print('These statements will be')
  print('executed when the conditon is satisfied.')
else:
  print('These statements will be')
  print('executed when the conditon is not satisfied.')
print('This is the next statement.')
````

### When the number of cases is 3 or more

```
if condition 1 :
  statements 1
elif condition 2 :
  statements 2
elif condition 3 :
  statements 3
....
elif condition n :
  statements n
else :
  statements n+1
```
* There are only two boolean values, `True` or `False`, so if you want to divide into three or more cases, you need to combine multiple conditions.
* Check conditions 1,2,3, ..., n in order, and perform the statements for the condition that first holds.
* You can write as many `elif` as you like.


````python
if 1 == 1 :
  print('These statements will be')
  print('executed when conditon 1 is satisfied.')
elif 1 == 1 :
  print('These statements will be')
  print('executed when conditon 1 is not satisfied and condition 2 is satisfied.')
else:
  print('These statements will be')
  print('executed when neither conditon 1 nor condition 2 is satisfied.')
print('This is the next statement.')
````

* Assume that some numbers are assigned to variables `x`,` y`, and `z`.  Choose expressions that have the same meaning among the followings.
 * ğŸ”—[De Morgan's Laws - Brilliant Math & Science Wiki](https://brilliant.org/wiki/de-morgans-laws/)

1. `x < y < z`
2. `x < y and y < z`
4. `not(x >= y and y >= z)`
7. `not(x >= y) or not(y >= z)`

Answer:

````python
# è«–ç†çš„ãªç­‰ä¾¡æ€§ã‚’æ¤œè¨¼ã™ã‚‹ãŸã‚ã®ãƒ†ã‚¹ãƒˆã‚±ãƒ¼ã‚¹ã‚’å®šç¾©
# (x < y) ã¨ (y < z) ã®çµæœã«åŸºã¥ã„ã¦ã€AND/OR ã®çµæœãŒç•°ãªã‚‹ã‚±ãƒ¼ã‚¹ã‚’å«ã‚€

# å„ã‚±ãƒ¼ã‚¹ã§æœŸå¾…ã•ã‚Œã‚‹ (x < y) ã¨ (y < z) ã®çœŸå½å€¤ã®çµ„ã¿åˆã‚ã›:
# T: True, F: False
# Case 1: T and T (AND: True, OR: True)
# Case 2: T and F (AND: False, OR: True)
# Case 3: F and T (AND: False, OR: True)
# Case 4: F and F (AND: False, OR: False)

test_cases = [
    (1, 2, 3),    # Case 1: True and True
    (1, 10, 5),   # Case 2: True and False
    (10, 5, 15),  # Case 3: False and True
    (5, 5, 5)     # Case 4: False and False
]

# å…¨ã¦ã®ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆã‚’lambdaé–¢æ•°ã¨ã—ã¦å®šç¾©
statements = [
    lambda x, y, z: x < y < z,                   # 1. AND (åŸºæœ¬å½¢)
    lambda x, y, z: x < y and y < z,             # 2. AND
    lambda x, y, z: x < y or y < z,              # 3. OR (åŸºæœ¬å½¢)
    lambda x, y, z: not(x >= y) and not(y >= z), # 4. AND (notã¨>=ã«ã‚ˆã‚‹è¡¨ç¾)
    lambda x, y, z: not(x >= y) or not(y >= z),  # 5. OR (notã¨>=ã«ã‚ˆã‚‹è¡¨ç¾)
    lambda x, y, z: not(x >= y and y >= z),      # 6. OR (ãƒ‰ãƒ»ãƒ¢ãƒ«ã‚¬ãƒ³æ³•å‰‡: not(A and B) => not A or not B)
    lambda x, y, z: not(x >= y or y >= z)        # 7. AND (ãƒ‰ãƒ»ãƒ¢ãƒ«ã‚¬ãƒ³æ³•å‰‡: not(A or B) => not A and not B)
]

statement_labels = [
    "1. x < y < z",
    "2. x < y and y < z",
    "3. x < y or y < z",
    "4. not(x >= y) and not(y >= z)",
    "5. not(x >= y) or not(y >= z)",
    "6. not(x >= y and y >= z)",
    "7. not(x >= y or y >= z)"
]

# çµæœã‚’ä¿å­˜ã™ã‚‹ãŸã‚ã®è¾æ›¸
results = {label: [] for label in statement_labels}

# å„ãƒ†ã‚¹ãƒˆã‚±ãƒ¼ã‚¹ã§å…¨ã¦ã®ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆã‚’è©•ä¾¡
for x, y, z in test_cases:
    for i, stmt in enumerate(statements):
        result = stmt(x, y, z)
        results[statement_labels[i]].append(result)

print("--- è«–ç†çš„ç­‰ä¾¡æ€§ã®æ¤œè¨¼çµæœ ---")
print("ãƒ†ã‚¹ãƒˆã‚±ãƒ¼ã‚¹: (x, y, z)")

# çµæœã‚’è¡¨å½¢å¼ã§è¡¨ç¤º
header = ["ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆ"]
for x, y, z in test_cases:
    header.append(f"({x}, {y}, {z})")

print("-" * 70)
print(f"{header[0]:<30} | {header[1]:<10} | {header[2]:<10} | {header[3]:<10} | {header[4]:<10}")
print("-" * 70)

# ã‚°ãƒ«ãƒ¼ãƒ—A (AND)
and_results = []
# ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆ1, 2, 4, 7ãŒANDã‚°ãƒ«ãƒ¼ãƒ—
and_indices = [0, 1, 3, 6] 
for i in and_indices:
    label = statement_labels[i]
    row = [label] + results[label]
    and_results.append(results[label])
    print(f"{row[0]:<30} | {str(row[1]):<10} | {str(row[2]):<10} | {str(row[3]):<10} | {str(row[4]):<10}")

# ã‚°ãƒ«ãƒ¼ãƒ—Aã®ç­‰ä¾¡æ€§ãƒã‚§ãƒƒã‚¯
# all(r == and_results[0] for r in and_results) ã¯ã€æœ€åˆã®ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆã®çµæœã¨ä»–ãŒå…¨ã¦ä¸€è‡´ã™ã‚‹ã‹ã‚’ãƒã‚§ãƒƒã‚¯
is_group_a_equal = all(r == and_results[0] for r in and_results)
print("-" * 70)
print(f"**ã‚°ãƒ«ãƒ¼ãƒ—A (AND) ã®æ¤œè¨¼çµæœ**: {'OK (å…¨ã¦ã®ã‚±ãƒ¼ã‚¹ã§ä¸€è‡´)' if is_group_a_equal else 'NG (ä¸ä¸€è‡´ã‚ã‚Š)'}")
print("-" * 70)

# ã‚°ãƒ«ãƒ¼ãƒ—B (OR)
or_results = []
# ã‚¹ãƒ†ãƒ¼ãƒˆãƒ¡ãƒ³ãƒˆ3, 5, 6ãŒORã‚°ãƒ«ãƒ¼ãƒ—
or_indices = [2, 4, 5]
for i in or_indices:
    label = statement_labels[i]
    row = [label] + results[label]
    or_results.append(results[label])
    print(f"{row[0]:<30} | {str(row[1]):<10} | {str(row[2]):<10} | {str(row[3]):<10} | {str(row[4]):<10}")

# ã‚°ãƒ«ãƒ¼ãƒ—Bã®ç­‰ä¾¡æ€§ãƒã‚§ãƒƒã‚¯
is_group_b_equal = all(r == or_results[0] for r in or_results)
print("-" * 70)
print(f"**ã‚°ãƒ«ãƒ¼ãƒ—B (OR) ã®æ¤œè¨¼çµæœ**: {'OK (å…¨ã¦ã®ã‚±ãƒ¼ã‚¹ã§ä¸€è‡´)' if is_group_b_equal else 'NG (ä¸ä¸€è‡´ã‚ã‚Š)'}")
print("-" * 70)

# çµè«–
print("\nçµè«–:")
if is_group_a_equal and is_group_b_equal:
    print("æŒ‡å®šã•ã‚ŒãŸå…¥åŠ›å€¤ã«ãŠã„ã¦ã€è«–ç†çš„ã«ç­‰ä¾¡ãªã‚°ãƒ«ãƒ¼ãƒ—åˆ†ã‘ãŒè¨¼æ˜ã•ã‚Œã¾ã—ãŸã€‚")
else:
    print("è«–ç†çš„ãªç­‰ä¾¡æ€§ã®æ¤œè¨¼ã«å¤±æ•—ã—ã¾ã—ãŸã€‚")
````
