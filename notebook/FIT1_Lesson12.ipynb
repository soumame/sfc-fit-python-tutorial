# Lesson12 Data Structures



* Some data handled by Python cannot be decomposed into components such as numbers and characters, and some can be decomposed into components such as strings and range types.
* There are many other data types that have a structure (composed of multiple elements), but this time we will study the *list* type and the *dictionary* type.

## List type

üîó[Python Lists - w3schools](https://www.w3schools.com/python/python_lists.asp#:~:text=%2C%20%22cherry%22%5D-,List,with%20different%20qualities%20and%20usage.)

* The range type is a regular sequence of numbers, while the list type is a list of arbitrary data.
* To create a list (list type data), arrange the element data in square brackets.

```python
x = ['Alice', 'Bob', 'Carol']
print(x)
```

* Elements can be a mixture of different types of data.

```python
x = ['Alice', 18, True]
print(x)
```

* A list can have 0 elements.

```python
x = []
print(x)
```

* You can use the `list` function to convert a multi-element data (such as a string or a range) into a list.

```python
print(list(range(10)))
print(list(range(1,10,2)))
print(list('Hello, world!'))
```

### Extraction by a for-statement

* You can take out the elements one by one with a for-statement and process them.

```python
for n in  ['Alice', 'Bob', 'Carol'] :
  print('Hello, ' + n)
```

### Extraction by specifying an index

* An element can be taken out by specifying its index (position of the element)
 * Note that the index starts at 0.
* Specify the index in square brackets after the list.

```python
x = ['Alice', 'Bob', 'Carol']
print(x[0])
print(x[2])
```

### Extraction of multiple elements by specifying a slice

üîó[How to slice a list, string, tuple in Python - note.nkmk.me](https://note.nkmk.me/en/python-slice-usage/)

* You can take out multiple elements at once in the form of "from what number to what number".
* Write `start : end : step` in the square brackets after the list.
 * Extract elements every `step` position from the index `start` and before the index `end`. Note that the position of `end` is not included.
 * If `step` is omitted, it will be 1.
* This kind of specification of element positions is called a *slice*.


```python
x = ['Alice', 'Bob', 'Carol', 'Dave', 'Ellen']
print(x[0:5:2])
print(x[1:4])
```

* Slices can be used not only for lists but also for strings and ranges.

```python
x = 'Hello, world!'
print(x[0:5])
```

### Changing elements

* You can specify an index and assign a new value to that location.

```python
x = ['Alice', 'Bob', 'Carol']
x[2] = 'Charlie'
print(x)
```

## Dictionary type

üîó[Python Dictionaries - w3scools](https://www.w3schools.com/python/python_dictionaries.asp)

* The dictionary type is for storing the correspondence between data.
* To create a dictionary (dictionary type data), arrange `data: data` in curly braces separated by commas.
 * The left side of `:` is called the *key*, and the right side is called the *value*.

```python
x = {'dog':'perro', 'cat':'gato', 'horse':'caballo'}
print(x)
```

### Extraction by a for-statement

* You can take out the keys one by one by using a for-statement.
 * The order in which they are taken out is not necessarily the order in which they are written.

```python
x = {'dog':'perro', 'cat':'gato', 'horse':'caballo'}
for k in x :
  print(k)
```

### Extraction by specifying a key

* You can specify a key to take out the corresponding value.
* Specify the key in square brackets after the dictionary.

```python
x = {'dog':'perro', 'cat':'gato', 'horse':'caballo'}
print(x['cat'])
```

* Write a program where a computer and a human play rock-paper-scissors.
 * Enters the human hand from the keyboard.
 * The computer chooses its hand randomly using the `choice` function of the `random` module.
   * üîó[Random sampling from a list in Python (random.choice, sample, choices) - note.nkmk.me](https://note.nkmk.me/en/python-random-choice-sample-choices/)
 * Shows both hands and which one wins (or draws).

```python
hands = ['rock', 'scissors', 'paper']
#win rule
rule = { 
    hands[0]: hands[1], #rock beats scissors
    hands[1]: hands[2], #scissors beats paper
    hands[2]: hands[0]  #paper beats rock
}

print('Rock, scissors, paper...')

human = input()

#validation
if human not in rule :
    print('Please enter a valid hand')
else:
    import random
    computer = random.choice(hands)
    print('You:', human, 'Computer:', computer)
    if human == computer :
        print('Draw')
    else:
        if rule[human] == computer :
            print('You win')
        else:
            print('You lose')
```

„ÄêExercise 10-11 ‚òÜ‚òÜ‚òÜ‚òÜ‚òÜ„Äë
* Write a program that enters the number of days of the year and shows what month and day it will be (in a normal year).
 * Use a list of days in each month and a while-statement.

```python
days = int(input('How many days?'))
year = int(input('What year?'))
defaultDays = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

months = {
    0: 'January',
    1: 'February',
    2: 'March',
    3: 'April',
    4: 'May',
    5: 'June',
    6: 'July',
    7: 'August',
    8: 'September',
    9: 'October',
    10: 'November',
    11: 'December'
}

def isLeapYear(year) :
    if year % 400 == 0 :
        return True
    if year % 100 == 0 :
        return False
    if year % 4 == 0 :
        return True

# in case of leap year, February has 29 days
if isLeapYear(year) :
    defaultDays[1] = 29

# find the month, and the remaining days
month = 0
while days > 0 :
    if days - defaultDays[month] > 0 :
        days -= defaultDays[month]
        month += 1
    else:
        break

print('It is ' + months[month] + ' ' + str(days) + 'th')


```
