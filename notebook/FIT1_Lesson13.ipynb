# Lesson13 Function Definitions

* There are built-in functions such as `len` and` print`, but you can also create new ones by yourself.
* Creating a function has the following merits:
 * When you do the same thing many times, you only have to write it once.
 * When the program becomes long, it can be divided into several functions to make it easier to read and write.


## How to write function definitions

üîó[Python Functions - w3schools](https://www.w3schools.com/python/python_functions.asp#:~:text=A%20function%20is%20a%20block,return%20data%20as%20a%20result.)

* Statemetns to be executed when a function is called is the *definition* of that function.

```
def function_name (parameters)„ÄÄ:
  statements
```
* As for the function name, you can use alphabets, numbers, and underscores in the same way as variable names.
* The parameters are variables to receive the data.
 * If there are more than one, separate them with commas.
 * They do not have to be, but the parentheses cannot be omitted.
* You must deepen the indentation for the statements part.
* When you execute the function definition, it becomes possible to use the function with that name.
 * The statements part is not executed yet.
 * Calling that function elsewhere in the program will execute the statements part.


````python
def hello() :
  print('Hello, world!')
````

````python
hello()
````

## Passing data to functions

* When calling a function, we write data in the following parenthesis.  They are called arguments.
* Arguments are assigned to the parameters of the function definition.
 * Often the terms *parameter* and *argument* are used synonimously.  To clearly distinguish, they are called *formal parameter* and *actual parameter*.

````python
def hello_someone(name) :
  print('Hello, ' + name + '!')
````

````python
hello_someone('Alice')
hello_someone('Bob')
````

### Postional parameters

* When there are multiple parameters, if nothing is specified, arguments will be matched in order.

````python
def meets(name1, name2) :
  print(name1 + ' meets ' + name2 + '.')
````

````python
meets('Alice', 'Bob')
````

* The number of arguments and parameters must be the same.

````python
meets('Alice')
````

## Scope of variables

* Each variable has its *scope*.
* Variables with the same name but with different scopes are different variables.

### Parameters

* A parameter used in a function definition is a kind of variable, and its scope is inside the function.

````python
def plus_two(x):
  x += 2
````

````python
def plus_three(x):
  x += 3
````

````python
x = 1
plus_two(x)
print(x)
plus_three(x)
print(x)
````

### Local variables

* A variable assigned in a function definition is called a *local variable*.
* The scope of a local variable is inside the function.

````python
def assign_two():
  x = 2
````

````python
def assign_three():
  x = 3
````

````python
x = 1
assign_two()
print(x)
assign_three()
print(x)
````

### Gloable variables

* A variable assigned outside function definitions is called a *global variable*.
* The scope of a global variable is the entire program. However, it cannot be assigned in function definitions.
 * If you assign a global variable in a function definition, a new local variable with the same name will be created.

````python
def print_global():
  print(x)
````

````python
x = 1
print_global()
x = 2
print_global()
````

### Why you need a scope?

* When writing a large program by multiple people, it is possible to accidentally use the same variable name.
 * If global variables are used, the same variable will be processed differently, resulting in a messy result.
 * If local variables are used, even if you use the same name, they will be different variables, so they will not affect each other.
* You can call the function again while it is executing. This is called a *recursive call*.
 * If you only have global variables, the processing of the first call (still running) and the second call will be mixed.
 * If you use local variables, they will be different variables for each call, so the first call and the second call can be processed independently.

* Modify Exercise 10-05 and write a function that takes the number of months as an argument and returns the number of days in that month.

````python
def month_days(month):
    d31_months = [1,3,5,7,8,10,12]
    d30_months = [4,6,9,11]
    d28_months = [2]

    if month in d31_months :
        return 31
    elif month in d30_months :
        return 30
    elif month in d28_months :
        return 28
    else:
        return 'Hey, there is no such month!'
````

````python
month = int(input('What month?'))

print(month_days(month))
````

* Write a function that takes one integer greater than or equal to 2 as an argument and returns `True` if it is a prime number and` False` if it is not.
 * Check if the argument is divisible by a number in order from 2, and return `False` as soon as it is divisible.

````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(int(input('Enter a number: '))))
````
